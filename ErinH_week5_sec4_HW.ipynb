```python
# Loop over the following string to (1) count all the characters in the string and (2) 
# print out all the vowels -- "The quick brown fox jumps over the lazy dog"

loop1 = "The quick brown fox jumps over the lazy dog"

count = 0
for character in loop1:
    count += 1
print("Printing # of characters in the string...", count)

loop1 = "The quick brown fox jumps over the lazy dog"
vowels = "aeiou"

for character in loop1:
    if character in vowels:
        print("Printing all the vowels...", character)
```

```python
# Use range as a loop to calculate the sum of all the numbers from 1 to 100

total_sum = 0

for number in range(1, 101):
    total_sum += number 

print("Printing the sum of all the numbers from 1 to 100 is..", total_sum)
```
